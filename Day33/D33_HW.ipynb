### Q2:以上圖資料，計算當你看到長髮時，是女生的機率?

```python
# 男 90 (長9 短81) 
# 女 10 (長5 短 5)
# 長 14
# 短 86
```

```python
# P(A|B) = P(女生|長髮) = P(A&B)/P(B)
# method1
P_B = 14/100
P_A_B = 5/100
P_A_given_B = P_A_B / P_B
P_A_given_B
```

```python
# method2
def bayes(p_b_given_a, p_a, p_b_given_a_com):
  p_a_com = 1 - p_a
  return (p_b_given_a * p_a) / (p_b_given_a * p_a + p_b_given_a_com * p_a_com )

P_A = 10 / 100
P_B_given_A = P_A_B / P_A 
P_B_A_com = 9 / 100
P_B_given_A_com = P_B_A_com / (1 - P_A)
P_A_given_B = bayes(P_B_given_A, P_A, P_B_given_A_com)
P_A_given_B
```
