### 2. 如何在不用迴圈的情況下計算 (A+B)*(-A/2) ？那用迴圈怎麼做？


````python
# 記得先 Import 正確的套件
import numpy as np
````

````python
A = np.ones(3)*1
B = np.ones(3)*2
# without for loop -> broadcasting
ans = (A + B) * (-A / 2)
ans
````

````python
# using for-loop
ans = np.ones(A.size)
C = np.arange(3)
for i, j, k in zip(A, B, C):
  ans[k] = (i + j) * (-i / 2)
ans
````
