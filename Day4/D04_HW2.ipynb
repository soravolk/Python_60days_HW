```
#1.正常的談話的聲壓為20000微巴斯卡，請問多少分貝?
#請寫下程式
V1 = 20000
V0 = 20  # 計算分貝時，20微巴斯卡為參考值
decibel = 20 * np.log10(V1/V0)
decibel
```

```
#2.30分貝的聲壓會是50分貝的幾倍?
#公式移項過後可以得到 V1 = ?
#請寫下程式
V1_1 = V0 * 10**(30 / 20)
V1_2 = V0 * 10**(50 / 20)

ans = V1_1 / V1_2
ans
```
