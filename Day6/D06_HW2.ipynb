### 1. [簡答題] 請問下列這三種方法有什麼不同？

```
print(a.sum()) 
print(np.sum(a))
print(sum(a)) 
```



````python
import numpy as np

a = np.arange(10)
print(a.sum())
print(a.sum().dtype)
print(a)
print(np.sum(a))
print(np.sum(a).dtype)
print(a)
print(sum(a)) 
print(sum(a).dtype)
print(a)
````

相同，皆不會影響原陣列且回傳int64 </br>
其中sum(a)的sum()為python內建的函數

### 2. 請對一個 5x5 的隨機矩陣作一般化的操作。


````python
A = np.random.random((5, 5)) * 7
print(A)
# normalize到0 ~ 1的區間
A = (A - A.min()) / (A.max() - A.min())
print("\nNormalization:")
print(A)
````

### 3. 請建立一個長度等於 10 的正整數向量，並且將其中的最大值改成 -1。


````python
a = np.arange(10)
print(a)
a[a.argmax()] = -1
print(a)
````
