### 2. 請撰寫一個判斷 a 的元素是否等於指定資料型態的函式，並且通過測試。



````python
import numpy as np;
````

````python
def is_dtype(a, t):
      if (a.dtype is t):
        return True
      else:
        return a.dtype is np.dtype(t)
         
````

````python
a = np.random.randint(10, size=6) 

# 測試一
print(is_dtype(a, 'int')) # True

# 測試二
print(is_dtype(a, np.int)) # True

# 測試三
print(is_dtype(a, np.dtype('int'))) # True
````
