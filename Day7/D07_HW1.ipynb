### 1. 請比較對一個 1000 x 100 x 100 的陣列，使用不同方法對每一個元素 +1 的時間比較。




```python
import numpy as np
```

```python
Z = np.random.randint(0, 10, 1000000).reshape(100, 100, 100)
```

```python
%%timeit -n 10 a = 2
for axis0 in Z:
  for row in axis0:
    for element in row:
      element = element + 1
```

```python
%timeit [Z.reshape(1, -1) + 1]
```

```python
%%timeit -n 10 a = 2
for a in Z.flat:
  a = a + 1
```

```python
%%timeit -n 10 a = 2
for a in np.nditer(Z):
  a = a + 1
```

### 2. 如何從一個陣列中，找出出現頻率最高的數值與位置？


```python
Z = np.random.randint(0,10,50)
print(Z)
```

```python
loc = np.bincount(Z.ravel()).argmax()
ans = np.array(np.where(Z == loc))
print("The most frequent value: ", loc)
print("location: ", ans)
```

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


```python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))


b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))


c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
```

```python
print(list(a))
```

```python
for x in list(b):
  print(list(x))
```

```python
for x in list(c):
  for y in list(x):
   print(list(y))
```
