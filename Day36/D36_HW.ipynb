## 作業:
某工廠想知道兩條不同的生產線的產品不良率是否有所不同，由兩條生產線中各抽取 300 個樣本，第一組有  75 個不良品，第二組有 30 個不良品，我們可以宣稱生產線所生產出的產品不良率不相同? (以 0.05 為顯著水準)?    
(提示:透過課程投影片的步驟，需思考 𝐻_0、 𝐻_1 的寫法和範例不同唷。)   



第一條生產線的不良率$p_a$；第二條生產線的不良率$p_b$  
$H_0$: $p_a$ = $p_b$  
$H_1$: $p_a$ $\neq$ $p_b$

```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
import statsmodels.stats.proportion as test
A = [75, 30]
B = [300, 300]
test.proportions_ztest(A, B, alternative='two-sided')
```

顯著水準為0.05，拒絕 $H_0$。  
兩條產線的不良率不同。
