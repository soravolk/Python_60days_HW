```python
# method1
x = x.copy()
c = x.loc[:,'Sex'].astype('category')
d = dict(enumerate(c.cat.categories))
x['Sex'] = x['Sex'].astype('category').cat.codes
```

```python
c = x['Embarked'].astype('category')
d = dict(enumerate(c.cat.categories))
x['Embarked'] = x['Embarked'].astype('category').cat.codes
```
