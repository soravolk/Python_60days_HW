```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
## 輸入資料
boy = np.array([164,175,183,173,176,159,165,165,169,151,156,
                 163,169,144,170,177,165,160,164,171])
girl = np.array([169,170,162,154,183,173,169,167,170,185,162,
                  175,168,151,181,170,182,156,159,160])
```

```python
#Q1:試著用今天所教的內容，如何描述這兩組資料的樣態?
#暗示:  分別計算男生和女生的統計量
mean = [np.mean(boy), np.mean(girl)]
median = [np.median(boy), np.median(girl)]
skew = [stats.skew(boy), stats.skew(girl)]
kurtosis = [stats.kurtosis(boy), stats.kurtosis(girl)]
```

```python
plt.hist(boy, alpha=.4, bins=50, color='y')
plt.title('boy, skewness={}, kurtosis={}'.format(round(skew[0],2), round(kurtosis[0],2)))
plt.axvline(x=mean[0])
plt.axvline(x=median[0], color='black')
# 微右偏
```

```python
plt.hist(girl, alpha=.4, bins=50, color='y')
plt.title('girl, skewness={}, kurtosis={}'.format(round(skew[1],2), round(kurtosis[1],2)))
plt.axvline(x=mean[1])
plt.axvline(x=median[1], color='black')
# 微左偏
```

```python
#Q2: 請問男生和女生在平均身高上誰比較高? 
print(mean[0])
print(mean[1])
# 女生高
```
